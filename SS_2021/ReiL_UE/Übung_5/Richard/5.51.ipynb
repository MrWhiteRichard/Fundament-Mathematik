```python
import os

import pandas as pd
import matplotlib.pyplot as plt
```

```python
parameters = {
    'gamma': 0.9,
    'alpha': 0.1,
    'feature_info': {'mode': 'polynomials', 'degree': 1},
    'episode_number_max': 100
}
```

```python
pd.DataFrame([1, 2, 3]).to_csv('jojojo.csv')
```

```python
file_name = f'{parameters}.csv'.replace(':', ' =')
runs = pd.read_csv(file_name, index_col=0)
```

```python
plt.figure(
    figsize=(16, 8)
)

label = ''
label += r'$\gamma = ' + str(parameters["gamma"]) + '$' + ', '
label += r'$\alpha = ' + str(parameters["alpha"]) + '$' + ', '
label += 'feature mode: ' + parameters['feature_info']['mode'] + ', '
label += 'degree: ' + str(parameters['feature_info']['degree'])

runs.mean(axis=1).plot(label=label)

_, runs_amount = runs.shape

plt.xlabel('Episode')
plt.ylabel(f'$G_0$, Total reward on episode, averaged over {runs_amount} runs')

plt.legend()
plt.show()
```
