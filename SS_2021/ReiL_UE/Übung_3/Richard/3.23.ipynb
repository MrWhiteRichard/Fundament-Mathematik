```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda gamma: (56 * gamma + 4) / (gamma - 4)
```

```python
x = np.arange(0, 1, 0.01)
y = f(x)

fig = plt.figure(figsize = (15, 5))

plt.plot(x, y)

plt.xlabel(r'$\gamma$')
plt.ylabel(r'$v_\pi(15)$')
plt.grid(linestyle = ':')

plt.show()
```
