```python
from sympy import plot_implicit, symbols, Eq
```

```python
s_1, s_2 = symbols('s_1 s_2')
z = 1.959964
#s_1,s_2 = symbols('')
x_1 = 8.2
x_2 = 6.9
n_1 = 55
n_2 = 159
klam = ((x_1-x_2)/z)**2
a = n_1*klam
b = n_2*klam


p1 = plot_implicit(Eq(s_1**2/a + s_2**2/b,1),(s_1,-6.5,6.5),(s_2,-10.5,10.5),show = True)
p1.save('C:/Users/Salli/Documents/Fundament-Mathematik/SS_2021/EStat_UE/Übung_12/Christian_S/ell_2.png')
```
