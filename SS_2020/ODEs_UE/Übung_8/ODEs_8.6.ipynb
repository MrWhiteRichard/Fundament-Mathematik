```python
import sympy as sp
```

```python
t = sp.Symbol('t')

A = sp.Matrix([[-1 + 3/2 * sp.cos(t)**2,           1 - 3/2 * sp.sin(t) * sp.cos(t)],
               [-1 - 3/2 * sp.sin(t) * sp.cos(t), -1 + 3/2 * sp.sin(t)**2         ]])

lamda = sp.Symbol('lamda')
p = A.charpoly(lamda)

print("Charakteristisches Polynom ... normal ...")
display(p)

print("Charakteristisches Polynom ... faktorisiert ...")
p = sp.factor(p)
display(p)

print("Eigenwerte (nochmal) ...")
for eigen_value in A.eigenvals():
    display(eigen_value)
    print("Stimmt der Eigenwert mit dem Charakteristischen Polynom?")
    display(sp.simplify(p.subs(lamda, eigen_value)) == 0)
```

```python
y = sp.exp(t/2) * sp.Matrix([-sp.cos(t), sp.sin(t)])

lhs = y.diff(t)
lhs = sp.simplify(lhs)

rhs = A * y
rhs = sp.simplify(rhs)

display(sp.Eq(lhs, rhs))
```
