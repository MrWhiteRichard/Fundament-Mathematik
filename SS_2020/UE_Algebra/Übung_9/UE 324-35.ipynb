```python
import numpy as np
import pandas as pd

array = lambda n: np.array([[a** + 5 * b**2 for a in range(n)] for b in range(n)])
n = 8

array = array(n)
display(pd.DataFrame(array))
Truth = np.prod(array == 2) or np.prod(array == 3)

print("Is there a", 2, "or a", 3, "?")
display(Truth == True)
```
