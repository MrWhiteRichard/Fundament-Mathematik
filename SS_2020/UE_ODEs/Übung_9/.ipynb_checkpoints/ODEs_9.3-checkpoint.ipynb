```python
import sympy as sp
```

```python
gamma = sp.Symbol('gamma')
lamda = sp.Symbol('lambda')
mu    = sp.Symbol('mu')

a = sp.Symbol('a')

x, y, z = sp.symbols('x   y   z')

stars = x_star, y_star, z_star = sp.symbols('x^* y^* z^*')

f = lambda x, y, z: sp.Matrix([-lamda*x*y - mu*x + mu*a, lamda*x*y - mu*y + gamma*y, gamma*y - mu*z])
V = lambda x, y, z: x - x_star*sp.ln(x) + y - y_star*sp.ln(y)
```

```python
gradient_V = sp.Matrix([V(x, y, z).diff(v) for v in [x, y, z]])
```

```python
Df = lambda x, y, z: sp.Matrix([f(x, y, z).diff(v).transpose() for v in (x, y, z)]).transpose()
```

```python
f(*stars)
```

```python
A = Df(*stars)
A
```
