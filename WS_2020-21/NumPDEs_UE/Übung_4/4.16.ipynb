```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def skizzieren(p):

    z = sum([
        [
            np.array([j / p, k / p])
            for j in range(p - k + 1)
        ]
        for k in range(p + 1)
    ], [])

    z = np.array(z).T

    fig = plt.figure(figsize = (8, 8))

    plt.suptitle(f'$p = {p}$')
    plt.scatter(*z)
    plt.grid(linestyle = ':')
    plt.show()
```

```python
for p in range(1, 4+1):
    skizzieren(p)
```
