```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def plot_distribution_function(function, jumps, interval):

    fig = plt.figure()

    plt.plot(
        interval[interval < jumps[0]],
        function(
            interval[interval < jumps[0]]
        )
    )

    for n in range(1, len(jumps)):
        plt.plot(
            interval[(jumps[n-1] <= interval) * (interval < jumps[n])],
            function(
                interval[(jumps[n-1] <= interval) * (interval < jumps[n])]
            )
        )

    plt.plot(
        interval[interval >= jumps[-1]],
        function(
            interval[interval >= jumps[-1]]
        )
    )

    plt.scatter(jumps, function(jumps), color = 'black')

    plt.grid(linestyle = ':')
    fig.show()
```

```python
@np.vectorize
def F(x):

    if x < 0:
        return 0
    elif 0 <= x < 1:
        return x
    elif 1 <= x < 2:
        return x + 1
    elif x >= 2:
        return 3
    else:
        return None

plot_distribution_function(
    F,
    np.array([0, 1, 2]),
    np.arange(-1, 3, 0.01)
)
```

```python
@np.vectorize
def G(x):
    if x < 1:
        return x
    elif 1 <= x < 2:
        return x ** 2
    elif x >= 2:
        return 5
    else:
        return None

plot_distribution_function(
    G,
    np.array([1, 2]),
    np.arange(-1, 3, 0.01)
)
```
