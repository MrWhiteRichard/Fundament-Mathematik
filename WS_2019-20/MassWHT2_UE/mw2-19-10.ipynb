```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D

mu_x = 0
variance_x = 5

mu_y = 0
variance_y = 5

a = -10
b = 10
dt = 0.01

x = np.arange(a, b, dt)
y = np.arange(a, b, dt)
X, Y = np.meshgrid(x, y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y
rv = multivariate_normal([mu_x, mu_y], [[variance_x, 0], [0, variance_y]])

fig = plt.figure(figsize = (16, 9))
ax = fig.gca(projection = '3d')
ax.plot_surface(X, Y, rv.pdf(pos), cmap = 'RdYlGn')

plt.show()
```

```python
import numpy as np
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D

c = 1
R = 2


fig = plt.figure(figsize = (16, 9))
ax = fig.gca(projection = "3d")

dt = 0.01

r = np.arange(0, R, dt)
phi = np.arange(0, 2*np.pi, dt)

r, phi = np.meshgrid(r, phi)

X = r * np.cos(phi)
Y = r * np.sin(phi)
Z = c * np.exp(-(X**2 + Y**2))

xmas = "gold"
ax.plot_surface(X, Y, Z, color = xmas)

plt.show()
```
