```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = 0.001
b = 100
dt = 0.1

grid = np.arange(a, b, dt)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10, 5))

ax1.set_title("$t \cdot \cos(1/t)$")
ax2.set_title("$t \cdot \sin(1/t)$")

ax1.plot(grid, grid * np.cos(1/grid))
ax1.grid(linestyle = ':')

ax2.plot(grid, grid * np.sin(1/grid))
ax2.grid(linestyle = ':')

plt.show()
```
